```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import preprocess_input
import tensorflow as tf
```

```python
# Load VGG19 model with pre-trained ImageNet weights
base_model = VGG19(weights='imagenet', include_top=False)

# Extract output of the relu2_1 (block2_conv1 layer followed by ReLU activation)
model = Model(inputs=base_model.input, outputs=base_model.get_layer('block3_conv1').output)
```

```python
# Load and preprocess the input image
img_path = 'style.jpg'  # Replace with your image path
img = image.load_img(img_path, target_size=(224, 224))
img_data = image.img_to_array(img)
img_data = np.expand_dims(img_data, axis=0)
img_data = preprocess_input(img_data)
```

```python
# Extract features from the relu2_1 layer
features = model.predict(img_data)

# Limit the number of feature maps to 200 (if available)
num_feature_maps = min(features.shape[-1], 200)
```

```python
def gram_matrix(x):
    # Reshape the feature maps
    x = tf.reshape(x, (tf.shape(x)[-1], -1))  # Shape: (num_channels, height * width)
    return tf.matmul(x, x, transpose_b=True)  # Shape: (num_channels, num_channels)
```

```python
# Calculate Gram matrices for each layer
gram_matrices = [gram_matrix(feature) for feature in features]
```

```python
def plot_gram_matrix(gram, layer_name):
    plt.figure(figsize=(8, 8))
    plt.imshow(gram, cmap='viridis', interpolation='nearest')
    plt.title(f'Gram Matrix for {layer_name}')
    plt.colorbar()
    plt.axis('off')
    plt.show()
```

```python
for i, gm in enumerate(gram_matrices):
    # Convert to a 2D numpy array for visualization
    gm_2d = gm.numpy()
    plot_gram_matrix(gm_2d, f'Layer block3_conv1')
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image as keras_image
```

```python
# Load and preprocess images
def load_and_preprocess_image(img_path):
    img = keras_image.load_img(img_path, target_size=(512, 512))
    img = keras_image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = tf.keras.applications.vgg19.preprocess_input(img)
    return img
```

```python
def deprocess_image(img):
    img = img.squeeze()
    img = img + 103.939, img + 116.779, img + 123.68  # VGG normalization
    img = np.clip(img, 0, 255).astype('uint8')
    return img
```

```python
# Define the model
def get_model():
    vgg = tf.keras.applications.VGG19(include_top=False, weights='imagenet')
    # Specify content layer and style layers
    content_layer = 'block5_conv2'  # Content layer
    style_layers = [
        'block1_conv1', 
        'block2_conv1', 
        'block3_conv1', 
        'block4_conv1', 
        'block5_conv1'
    ]
    
    # Create a list of outputs from the model
    outputs = [vgg.get_layer(content_layer).output]
    for layer in style_layers:
        outputs.append(vgg.get_layer(layer).output)
    
    return tf.keras.Model(inputs=vgg.input, outputs=outputs)
```

```python
# Compute the Gram matrix
def gram_matrix(tensor):
    channels = int(tensor.shape[-1])
    a = tf.reshape(tensor, [-1, channels])
    g = tf.matmul(a, a, transpose_a=True)
    return g / tf.cast(tf.shape(a)[0] * channels, tf.float32)
```

```python
# Compute content and style loss
def content_loss(base_content, target):
    return tf.reduce_mean(tf.square(base_content - target))

def style_loss(base_style, target):
    return tf.reduce_mean(tf.square(gram_matrix(base_style) - gram_matrix(target)))

# Combine losses
def total_loss(base_content, base_style, target_content, target_style, alpha=1e-3, beta=1e-1):
    c_loss = content_loss(base_content, target_content)
    s_loss = style_loss(base_style, target_style)
    return alpha * c_loss + beta * s_loss
```

```python
# Load images
content_img = load_and_preprocess_image('image.jpg')
style_img = load_and_preprocess_image('style.jpg')
```

```python
# Get model
model = get_model()
content_features = model(content_img)[0]  # Content features
style_features = model(style_img)[1:]  # Style features
```

```python
# Initialize generated image
generated_img = tf.Variable(content_img)

# Optimization loop
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
epochs = 1000
```

```python
# Step 1: Get the model
model = get_model()

# Step 2: Generate features for content and style images
outputs_content = model(content_img)
outputs_style = model(style_img)

# Step 3: Extract content and style features
content_features = outputs_content[0]  # Assuming first output is content
style_features = outputs_style[1:]      # Assuming remaining outputs are style features

# Step 4: Define layer names for later use
layer_names = [layer.name for layer in model.layers]

# Step 5: Training loop
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        generated_features = model(generated_img)
        
        # Calculate loss
        loss = total_loss(
            generated_features[0],  # Content features from generated image
            style_features[0],      # Adjust based on your style feature needs
            content_features,
            {layer_names[i]: generated_features[i] for i in range(len(generated_features))}
        )
    
    # Compute gradients and apply updates
    grads = tape.gradient(loss, generated_img)
    optimizer.apply_gradients([(grads, generated_img)])
    
    # Print loss every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch: {epoch}, Loss: {loss.numpy()}")

# Step 6: Convert and display the generated image
result_img = deprocess_image(generated_img.numpy())
plt.imshow(result_img)
plt.axis('off')
plt.show()
```
